```python
import os
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
# read the results from the file
with open('./results/conf_1_results.json','r') as f:
    results = json.load(f)
    train_loss_inc = np.array(results['train_loss_inc'])
    train_loss = np.array(results['train_loss'])
    valid_loss = np.array(results['valid_loss'])
    train_acc_inc = np.array(results['train_acc_inc'])
    train_acc = np.array(results['train_acc'])
    valid_acc = np.array(results['valid_acc'])

# plot the results 
fig, ax = plt.subplots(2,1,figsize=(10,10))
ax[0].plot(train_loss_inc, label='train loss inc')
ax[0].plot(train_loss, label='train loss', linestyle='--')
ax[0].plot(valid_loss, label='valid loss', linestyle='-.')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')

ax[1].plot(train_acc_inc, label='train acc inc')
ax[1].plot(train_acc, label='train acc', linestyle='--')
ax[1].plot(valid_acc, label='valid acc', linestyle='-.')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Accuracy')

ax[0].legend()
ax[1].legend()
plt.show()
```
