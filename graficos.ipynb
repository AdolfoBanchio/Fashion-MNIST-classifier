```python
import os
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
for i in range(1, 5):
    # Read the results from each configuration
    with open(f'./results/conf_{i}_results.json', 'r') as f:
        results = json.load(f)
        param = results['params']
        train_loss_inc = np.array(results['train_loss_inc'])
        train_loss = np.array(results['train_loss'])
        valid_loss = np.array(results['valid_loss'])
        train_acc_inc = np.array(results['train_acc_inc'])
        train_acc = np.array(results['train_acc'])
        valid_acc = np.array(results['valid_acc'])
    
    # Print the results
    print(f'Configuration {i}, Learning rate: {param["learning_rate"]}, Batch size: {param["batches_size"]}')  
    print('Last train loss: ',train_loss[-1])
    print('Last valid loss: ',valid_loss[-1])
    print('last train during training loss: ',train_loss_inc[-1])
    print('Last train accuracy: ',train_acc[-1])
    print('Last valid accuracy: ',valid_acc[-1])
    print('Last train during training accuracy: ',train_acc_inc[-1])
    print('=====================')
    # Plot the results
    fig = plt.figure(figsize=(10,10))
    gs = fig.add_gridspec(2, hspace=0)
    ax = gs.subplots(sharex=True)

    ax[0].plot(train_loss_inc, label='Train loss during training')
    ax[0].plot(train_loss, label='Train loss', linestyle='--')
    ax[0].plot(valid_loss, label='Valid loss', linestyle='-.')
    ax[0].set_ylabel('Loss')
    ax[0].legend()
    ax[0].set_title(f'Configuration {i} - Learning rate: {param["learning_rate"]}')
    ax[0].grid()

    ax[1].plot(train_acc_inc, label='Train accuracy during training')
    ax[1].plot(train_acc, label='Train accuracy', linestyle='--')
    ax[1].plot(valid_acc, label='Valid accuracy', linestyle='-.')
    ax[1].set_ylabel('Accuracy')
    ax[1].set_xlabel('Epoch')
    ax[1].legend()
    ax[1].grid()

    # Save the plot
    fig.savefig(f'./results/conf_{i}_results.png')
    plt.close(fig)

# Create the markdown file with the images
with open('./results/results.md', 'w') as md_file:
    for i in range(1, 5):
        # Read the learning rate from the results
        with open(f'./results/conf_{i}_results.json', 'r') as f:
            results = json.load(f)
            param = results['params']
        # Write the header
        md_file.write(f'### conf {i}, Lr={param["learning_rate"]}\n')
        # Insert the image
        md_file.write(f'![](./conf_{i}_results.png)\n\n')
```

```python
# accurracy of config 4

with open(f'./results/conf_4_results.json', 'r') as f:
    results = json.load(f)
    param = results['params']
    train_loss_inc = np.array(results['train_loss_inc'])
    train_loss = np.array(results['train_loss'])
    valid_loss = np.array(results['valid_loss'])
    train_acc_inc = np.array(results['train_acc_inc'])
    train_acc = np.array(results['train_acc'])
    valid_acc = np.array(results['valid_acc'])

# plot the accurracy of config 4
fig = plt.figure(figsize=(10,5))

plt.plot(train_acc_inc, label='Train accuracy during training')
plt.plot(train_acc, label='Train accuracy', linestyle='--')
plt.plot(valid_acc, label='Valid accuracy', linestyle='-.')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.grid()
plt.title(f'Configuration 4 - Learning rate: {param["learning_rate"]}, ADAM optimizer')
```
